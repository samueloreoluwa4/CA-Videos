## Information and Visualizations on Tags

```python
# All Tags per channel
ca['tags'].unique()
```

```python
# To separate all tag names from the channel tags
tgs = []
for i in ca['tags']:
    j = i.split('|') #Separating tag names at everypoint where there is '|'
    tgs.append(j) # appending tag names after spliting
tgs
```

```python
c = []
for a in tgs:
    for b in a:
        bb = b.strip('"')# deleting '"' from tag names
        if bb != '':
            c.append(bb)
c
```

```python
# Converting tag names to a dataframe after separating
d = pd.DataFrame(c)
d_sort = d.sort_values(by=0)
d_sort
```

```python
# To find the number of tag name in the dataset
d = d_sort.rename({0:'tags'}, axis='columns')
tg_count = d['tags'].value_counts()
tg_count[:10]
```

### Channels that Tagged Funny

```python
# Lets chech how many channels tagged funny
cca = ca[ca.tags.notnull()]
funny = np.where(cca['tags'].str.contains('funny'))
fun = funny[0]
len(fun)
```

```python
# To chech how many likes a channel that tagged funny gets
ch_likes = ca['likes'].loc[fun]
ch_likes = pd.DataFrame(ch_likes).reset_index()
ch_likes
```

```python
# Maximum number of likes gotten by a channel that tagged funny
Max_ch_likes = max(ch_likes['likes'])
Max_ch_likes
```

```python
sns.lineplot(x='index', y='likes', data=ch_likes)
plt.title('Plot of Likes for Channels that Tagged Funny')
plt.show()
```

### Channels that Tagged Comedy

```python
# How many channels tagged comedy
comm = np.where(cca['tags'].str.contains('comedy'))
comedy = comm[0]
len(comedy)
```

```python
# To chech how many views a channel that tagged funny gets
views = ca['views'].loc[comedy]
ch_views = pd.DataFrame(views).reset_index()
ch_views
```

```python
# Maximum number of views a channel that tagged comedy have
Max_ch_views = max(ch_views['views'])
Max_ch_views
```

```python
sns.lineplot(x='index', y='views', data=ch_views)
plt.title('Plot of Views for Channels that Tagged Comedy')
plt.show()
```
